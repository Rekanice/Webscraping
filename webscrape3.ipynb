```python
import requests
from bs4 import BeautifulSoup as bs

url = "https://www.scmp.com/news/world/united-states-canada/article/3095892/donald-trump-says-tiktok-will-be-out-business-us"

# html of the url
news = requests.get(url)

# to view the html
#news.text #as a string
#news.content

ptree = bs(news.content,'html.parser')
print(ptree.prettify())
```
